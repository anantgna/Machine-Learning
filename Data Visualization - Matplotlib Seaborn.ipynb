```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Line Plot**

```
plt.figure(figsize=(20,9))
plt.plot(boston_df['AGE'])
plt.title("Line Plot")
plt.xlabel("Row Indexes")
plt.ylabel("AGE")
plt.show()
```

```
plt.figure(figsize=(12,9))
plt.plot(boston_df['AGE'].sort_values(ascending=True).unique(),color='cyan')
plt.title("Line Plot",fontsize = 14)
plt.xlabel("Row Indexes")
plt.ylabel("AGE")
plt.show()
```

**Histogram Plot**

*   Continous Data Representation (Range vs Frequency of Data)



```
sns.set_style('darkgrid')
plt.figure(figsize=(15,10))
plt.hist(boston_df['MEDV'], bins = 20,color='cyan')
plt.title("Historgram Plot",fontsize = 14)
plt.xlabel("Range of MEDV")
plt.ylabel("Frequency (in Range)")
plt.show()
```

**Subplotting**

```
fig = plt.figure(figsize=(20,8))

# add subplot (1, 3, 1)(1 = Row, 3 = Column, 1 = position of image) 

ax = fig.add_subplot(1, 3, 1) 
sns.distplot(iris_df['sepal_length'], color ='c')
ax.set_title('Distribution  of Sepal Length')

ax = fig.add_subplot(1, 3, 2)
sns.distplot(iris_df['petal_length'], color ='m')
ax.set_title('Distribution  of Petal Length')

ax = fig.add_subplot(1, 3, 3)
sns.distplot(iris_df['petal_width'], color ='b')
ax.set_title('Distribution  of Petal Width')
```

**KDE(Kernal Density Estimator) Plot**

```
iris_setosa = iris_df.query("species == 'setosa'")
iris_virginica = iris_df.query("species == 'virginica'")
```

```
sns.kdeplot(iris_setosa['sepal_length'], shade = True, Label = 'setosa', color = 'red')
sns.kdeplot(iris_virginica['sepal_length'], shade = True, Label = 'virginica', color = 'blue')
```

```
sns.set_style('ticks')
sns.kdeplot(iris_setosa['sepal_length'], iris_setosa['sepal_width'], shade=True, Label ="Setosa", cmap = 'PuRd', shade_lowest = False)
plt.show()
```

**3-D Plot**

```
import numpy as np
import matplotlib.pyplot as plt
```

```
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection = '3d')
```

```
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection = '3d')

# Data for three dimensional

zpoints = np.linspace(0, 15, 1000)
xpoints = np.sin(zpoints)
ypoints = np.cos(zpoints)

ax.plot3D(xpoints, ypoints, zpoints, 'red')

ax.set_title("Line Plot")

plt.show()
```

```
def fun(x, y):
  return np.tan(np.sqrt(x**2 + y**2))
```

```
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)

X , Y = np.meshgrid(x, y)
Z = fun(X, Y)
```

```
plt.figure(figsize=(20,8))
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.contour3D(X, Y, Z, 50, cmap='Dark2')

plt.show()
```
